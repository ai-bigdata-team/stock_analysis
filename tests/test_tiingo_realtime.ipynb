```python
import websocket, json
import finnhub
import time
import threading
```

```python
finnhub_client = finnhub.Client(api_key="FINNHUB_API_KEY")
```

```python
symbols_data = finnhub_client.stock_symbols('US')
symbols = [item['symbol'] for item in symbols_data]
nasdaq_symbols = [item['symbol'] for item in symbols_data if item['mic'] in ['XNAS','XNGS']]
nyse_symbols = [item['symbol'] for item in symbols_data if item['mic'] == 'XNYS']

# symbols = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", 
#            "META", "NFLX", "NVDA", "JPM", "BAC"]
```

```python
print(len(nasdaq_symbols))
print(len(nyse_symbols))
```

```python
API_KEY = "TIINGO_API_KEY" # Tiingo API
```

```python
def on_message(ws, message):
    # print("Received:", message) # Không cần ghi ra file thì dùng dòng này thôi
    try:
        msg = json.loads(message)
        
        with open("test_messages.jsonl", "a") as f:
            f.write(json.dumps(msg) + "\n")
    except Exception as e:
        print("Error parsing message:", e)

def on_open(ws):
    # Gửi yêu cầu subscribe
    subscribe = {
        "eventName": "subscribe",
        "authorization": API_KEY,
        "eventData": {"tickers": nasdaq_symbols}
    }
    ws.send(json.dumps(subscribe))
    print("Subscribed to:", symbols)

    # Đặt hẹn giờ 10 phút để tự động đóng kết nối
    def stop_ws():
        print("Closing WebSocket after 10 minutes...")
        ws.close()

    timer = threading.Timer(600, stop_ws)  # 600 giây
    timer.start()

def on_error(ws, error):
    print("Error:", error)

def on_close(ws, close_status_code, close_msg):
    print("Connection closed")
```

```python
ws = websocket.WebSocketApp("wss://api.tiingo.com/iex",
                            on_message=on_message,
                            on_open=on_open,
                            on_error=on_error,
                            on_close=on_close)

ws.run_forever()
```
